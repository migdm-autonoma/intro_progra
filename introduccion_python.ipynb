## 3. 🔢 Tipos de datos

En Python hay distintos tipos de datos. Los más comunes son:

| Tipo | Nombre en Python | Ejemplo |
|------|-----------------|---------|
| Número entero | `int` | `42` |

````
entero  = 10
decimal = 3.14
texto   = "Python"
logico  = True

print(type(entero))
print(type(decimal))
print(type(texto))
print(type(logico))
````

## 4. ➕ Operaciones matemáticas

Python puede funcionar como una calculadora. Los operadores básicos son:

| Operador | Significado | Ejemplo |
|----------|------------|---------|
| `+` | Suma | `5 + 3` → `8` |
| `-` | Resta | `10 - 4` → `6` |
| `*` | Multiplicación | `3 * 7` → `21` |
| `/` | División | `15 / 4` → `3.75` |
| `**` | Potencia | `2 ** 8` → `256` |
| `//` | División entera | `15 // 4` → `3` |
| `%` | Módulo (resto) | `15 % 4` → `3` |


````
print(5 + 3)
print(10 - 4)
print(3 * 7)
print(15 / 4)
print(2 ** 8)
print(15 // 4)
print(15 % 4)
````

También puedes guardar el resultado en una variable:

````
precio = 1500
cantidad = 4
total = precio * cantidad

print("Total a pagar:", total, "pesos")
````

## 6. 🔀 Condicionales — `if`, `elif`, `else`

Los condicionales permiten que tu programa tome **decisiones**.

```
if condición:
    # código si la condición es verdadera
elif otra_condición:
    # código si la segunda condición es verdadera
else:
    # código si ninguna condición fue verdadera
```

> 🔎 La **indentación** (los espacios al inicio) es obligatoria en Python.


````
edad = int(input("Ingresa tu edad: "))

if edad >= 18:
    print("Eres mayor de edad ✅")
elif edad >= 13:
    print("Eres adolescente 🧑")
else:
    print("Eres niño/a 🧒")
````

Los operadores de comparación son:

| Operador | Significado |
|----------|-------------|
| `==` | igual a |
| `!=` | distinto de |
| `>` | mayor que |
| `<` | menor que |
| `>=` | mayor o igual |
| `<=` | menor o igual |


````
numero = int(input("Ingresa un número: "))

if numero % 2 == 0:
    print(numero, "es par")
else:
    print(numero, "es impar")
````

## 7. 🔁 Bucles — `for` y `while`

Los bucles repiten un bloque de código varias veces.

### 7.1 Bucle `for`
Se usa cuando sabes cuántas veces quieres repetir algo.


````
for i in range(5):
    print("Vuelta número", i)
````

`range(5)` genera los números del 0 al 4.  
Puedes personalizar el rango con `range(inicio, fin, paso)`:

````
# Contar del 1 al 10
for numero in range(1, 11):
    print(numero)
````

````
# Tabla de multiplicar del 7
for i in range(1, 11):
    print("7 x", i, "=", 7 * i)
````

### 7.2 Bucle `while`
Se repite **mientras** una condición sea verdadera:


````
contador = 0

while contador < 5:
    print("Contador:", contador)
    contador = contador + 1

print("¡Listo!")
````

## 8. 📋 Listas

Una **lista** guarda múltiples valores en una sola variable.  
Se define con corchetes `[ ]` y los elementos se separan por comas.


````
frutas = ["manzana", "naranja", "pera", "uva"]

print(frutas)
print("Primera fruta:", frutas[0])    # índice 0 = primer elemento
print("Última fruta:", frutas[-1])    # índice -1 = último elemento
print("Total de frutas:", len(frutas))
````

Puedes recorrer una lista con `for`:

````
notas = [6.5, 5.0, 7.0, 4.5, 6.0]

suma = 0
for nota in notas:
    suma = suma + nota

promedio = suma / len(notas)
print("Tu promedio es:", promedio)
````

## 9. 🛠️ Funciones

Una **función** es un bloque de código con nombre que puedes reutilizar.  
Se define con `def` y se llama escribiendo su nombre seguido de paréntesis.


````
def saludar(nombre):
    print("¡Hola,", nombre + "! 👋")

saludar("Ana")
saludar("Carlos")
saludar("Valentina")
````

````
def calcular_promedio(nota1, nota2, nota3):
    promedio = (nota1 + nota2 + nota3) / 3
    return promedio

mi_promedio = calcular_promedio(6.0, 5.5, 7.0)
print("Promedio:", mi_promedio)
````

## 🏆 Proyecto Final — Mini calculadora de notas

¡Ahora junta todo lo que aprendiste!  
Completa el programa para que calcule si el alumno aprueba o reprueba.


````
# 🎯 PROYECTO: Mini calculadora de notas

def calcular_promedio(notas):
    suma = 0
    for nota in notas:
        suma = suma + nota
    return suma / len(notas)

# Pedir notas al usuario
print("=== Calculadora de Notas ===")
nombre = input("Nombre del estudiante: ")

notas = []
for i in range(3):
    nota = float(input(f"Ingresa la nota {i+1}: "))
    notas.append(nota)

# Calcular promedio
promedio = calcular_promedio(notas)

# Mostrar resultado
print()
print("--- Resultado ---")
print("Estudiante:", nombre)
print(f"Promedio: {promedio:.1f}")

if promedio >= 4.0:
    print("Estado: ✅ APROBADO")
else:
    print("Estado: ❌ REPROBADO")
````
