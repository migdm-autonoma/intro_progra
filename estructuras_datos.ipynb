# -----------------------------------------
# PARTE 1: LISTAS
# -----------------------------------------

```python
print("=== LISTAS ===")

# 1. Crear una lista con 5 nombres
nombres = ["Ana", "Luis", "Pedro", "Sofía", "María"]

# 2. Agregar un nuevo nombre
nombres.append("Carlos")

# 3. Eliminar un nombre
nombres.remove("Pedro")

# 4. Imprimir la lista
print("Lista de nombres:", nombres)

# 🔹 DESAFÍOS LISTAS

# A. Ordenar la lista alfabéticamente
nombres.sort()
print("Lista ordenada:", nombres)

# B. Buscar si un nombre existe
buscar = "Ana"
if buscar in nombres:
    print(buscar, "está en la lista")
else:
    print(buscar, "NO está en la lista")

# C. Contar elementos
print("Cantidad de nombres:", len(nombres))
```

# -----------------------------------------
# PARTE 2: TUPLAS
# -----------------------------------------

```python
print("\n=== TUPLAS ===")

# 1. Crear una tupla con coordenadas (x, y)
coordenadas = (10, 20)

# 2. Mostrar valores
print("X:", coordenadas[0])
print("Y:", coordenadas[1])

# 🔹 DESAFÍO TUPLAS

# A. Crear una lista de tuplas con varias coordenadas
puntos = [(1,2), (3,4), (5,6)]

# B. Recorrer e imprimir
for punto in puntos:
    print("Punto:", punto)
```


# -----------------------------------------
# PARTE 3: DICCIONARIOS
# -----------------------------------------

```python
print("\n=== DICCIONARIOS ===")

# 1. Crear un diccionario con información de un estudiante
estudiante = {
    "nombre": "Ana",
    "edad": 20,
    "carrera": "Ingeniería"
}

# 2. Agregar un nuevo dato
estudiante["promedio"] = 6.5

# 3. Imprimir datos
print("Información del estudiante:")
for clave, valor in estudiante.items():
    print(clave, ":", valor)

# 🔹 DESAFÍOS DICCIONARIOS

# A. Modificar un valor
estudiante["edad"] = 21

# B. Eliminar un campo
del estudiante["carrera"]

print("\nDiccionario actualizado:", estudiante)

# C. Crear un diccionario con varios estudiantes
notas = {
    "Ana": 6.5,
    "Luis": 5.8,
    "Sofía": 6.2
}

# D. Calcular promedio
promedio = sum(notas.values()) / len(notas)
print("Promedio del curso:", promedio)
```

# -----------------------------------------
# PARTE 4: CONJUNTOS (SETS)
# -----------------------------------------

```python
print("\n=== SETS ===")

# 1. Crear un conjunto con números repetidos
numeros = {1, 2, 2, 3, 4, 4, 5}

# 2. Mostrar resultado
print("Conjunto:", numeros)

# 3. Agregar un número
numeros.add(6)

print("Conjunto actualizado:", numeros)

# 🔹 DESAFÍOS SETS

# A. Crear otro conjunto
otros_numeros = {4, 5, 6, 7, 8}

# B. Unión
print("Unión:", numeros.union(otros_numeros))

# C. Intersección
print("Intersección:", numeros.intersection(otros_numeros))

# D. Diferencia
print("Diferencia:", numeros.difference(otros_numeros))
```
