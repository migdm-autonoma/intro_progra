### 🔸 Ejercicio 6.4 — Juego de adivinanza con pistas ⭐⭐⭐

El programa elige aleatoriamente una palabra de una tupla. El usuario debe adivinarla letra por letra.  
En cada intento, muestra qué letras ya fueron adivinadas (usando un **set**) y el progreso de la palabra.

**Ejemplo:** Si la palabra es `"python"` y el usuario ya adivinó `p`, `t`, `h`:  
`Progreso: p _ t h _ _`


````python
# Escribe tu código aquí
import random

palabras = ("python", "lista", "tupla", "diccionario", "conjunto", "variable", "funcion")
palabra_secreta = random.choice(palabras)

letras_adivinadas = set()
intentos = 0
max_intentos = 8

print("=== ADIVINA LA PALABRA ===")
print(f"La palabra tiene {len(palabra_secreta)} letras.")

while intentos < max_intentos:
    # construye el progreso: muestra la letra si fue adivinada, _ si no
    progreso = ""
    for letra in palabra_secreta:
        if letra in letras_adivinadas:
            progreso += letra + " "
        else:
            progreso += "_ "
    
    print(f"\nProgreso: {progreso.strip()}")
    print(f"Letras usadas: {sorted(letras_adivinadas)}")
    print(f"Intentos restantes: {max_intentos - intentos}")
    
    # verifica si ya ganó (todas las letras adivinadas)
    if "_" not in progreso:
        print(f"\n🎉 ¡Ganaste! La palabra era '{palabra_secreta}'")
        break
    
    letra_ingresada = input("Adivina una letra: ").lower()
    
    if letra_ingresada in letras_adivinadas:
        print("Ya ingresaste esa letra.")
        continue
    
    letras_adivinadas.add(letra_ingresada)
    
    if letra_ingresada not in palabra_secreta:
        print(f"❌ La letra '{letra_ingresada}' no está en la palabra.")
        intentos += 1
else:
    print(f"\n💀 Se acabaron los intentos. La palabra era '{palabra_secreta}'")
````
