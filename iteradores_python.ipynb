## 1. ¿Qué es un bucle? 🔄 <a id='1'></a>

Un **bucle** (o *loop*) es una instrucción que le dice a Python:

> *"Repite este bloque de código varias veces"*

### Sin bucle — código repetitivo 😩

````python
# Sin bucle: hay que escribir cada línea a mano
print("Hola, estudiante 1")
print("Hola, estudiante 2")
print("Hola, estudiante 3")
print("Hola, estudiante 4")
print("Hola, estudiante 5")
````

### Con bucle — código elegante ✅

````python
# Con bucle: una sola instrucción hace lo mismo
for i in range(1, 6):
    print(f"Hola, estudiante {i}")
````

## 2. El bucle `for` 📋 <a id='2'></a>

El `for` recorre **cada elemento** de una secuencia (lista, string, rango, etc.).

### Sintaxis
```python
for variable in secuencia:
    # código que se repite
```

| Parte | Significado |
|-------|-------------|
| `variable` | Toma el valor de cada elemento en cada vuelta |
| `in` | Indica "dentro de" |
| `secuencia` | Lo que se va a recorrer |
| `:` | Inicio del bloque |
| Indentación (4 espacios) | Todo lo indentado se repite |

````python
# Ejemplo básico: recorrer una lista
frutas = ["manzana", "naranja", "pera", "uva"]

for fruta in frutas:
    print(f"🍎 Fruta: {fruta}")
````

````python
# El for puede hacer operaciones dentro del bucle
numeros = [10, 25, 7, 42, 3]

for numero in numeros:
    doble = numero * 2
    print(f"{numero} x 2 = {doble}")
````

## 3. La función `range()` 🔢 <a id='3'></a>

`range()` genera una **secuencia de números**. Es la compañera perfecta del `for`.

| Forma | Resultado | Ejemplo |
|-------|-----------|--------|
| `range(n)` | 0 hasta n-1 | `range(5)` → 0,1,2,3,4 |
| `range(inicio, fin)` | inicio hasta fin-1 | `range(2, 6)` → 2,3,4,5 |
| `range(inicio, fin, paso)` | con saltos | `range(0, 10, 2)` → 0,2,4,6,8 |

````python
# range(n) — del 0 al n-1
print("range(5):")
for i in range(5):
    print(f"  i = {i}")
````

````python
# range(inicio, fin) — del inicio al fin-1
print("Tabla de multiplicar del 3:")
for i in range(1, 11):
    print(f"  3 x {i} = {3 * i}")
````

````python
# range(inicio, fin, paso) — con saltos
print("Números pares del 0 al 20:")
for i in range(0, 21, 2):
    print(i, end=" ")

print()  # salto de línea

print("\nCuenta regresiva del 10 al 1:")
for i in range(10, 0, -1):
    print(i, end=" ")
````

## 4. Recorrer listas, strings, tuplas y diccionarios 🗂️ <a id='4'></a>

El `for` puede recorrer cualquier **iterable** en Python.

````python
# Recorrer un string — letra por letra
palabra = "Python"
print(f"Las letras de '{palabra}' son:")
for letra in palabra:
    print(f"  → {letra}")
````

````python
# enumerate() — obtener índice Y valor al mismo tiempo
colores = ["rojo", "verde", "azul", "amarillo"]

print("Lista de colores con su posición:")
for indice, color in enumerate(colores):
    print(f"  [{indice}] {color}")
````

````python
# Recorrer un diccionario
estudiante = {
    "nombre": "Ana",
    "edad": 20,
    "carrera": "Ingeniería",
    "promedio": 6.2
}

print("📋 Datos del estudiante:")
for clave, valor in estudiante.items():
    print(f"  {clave}: {valor}")
````

````python
# zip() — recorrer dos listas al mismo tiempo
alumnos   = ["Pedro", "María", "Luis", "Sofía"]
notas     = [6.5, 7.0, 5.8, 6.9]

print("📊 Calificaciones:")
for alumno, nota in zip(alumnos, notas):
    estado = "✅ Aprobado" if nota >= 6.0 else "❌ Reprobado"
    print(f"  {alumno}: {nota} → {estado}")
````

## 5. El bucle `while` ⏳ <a id='5'></a>

El `while` repite **mientras una condición sea verdadera**. Lo usas cuando no sabes exactamente cuántas veces repetir.

### Sintaxis
```python
while condicion:
    # código que se repite
    # ⚠️ debes actualizar la condición para no crear un bucle infinito
```

> ⚠️ **Cuidado con el bucle infinito**: si la condición nunca se vuelve `False`, el programa nunca termina.

````python
# Ejemplo básico de while
contador = 1

while contador <= 5:
    print(f"Vuelta número {contador}")
    contador += 1  # ← fundamental: actualiza el contador

print("¡Bucle terminado!")
````

````python
# Acumulador: sumar números hasta llegar a 100
suma = 0
numero = 1

while suma < 100:
    suma += numero
    numero += 1

print(f"La suma llegó a {suma} al sumar hasta el número {numero - 1}")
````

````python
# Simulación de menú interactivo (sin input real para el notebook)
opciones_simuladas = ["1", "2", "3", "salir"]
indice = 0

print("=== Simulación de menú ===")
opcion = ""

while opcion != "salir":
    opcion = opciones_simuladas[indice]
    indice += 1
    
    if opcion == "1":
        print("  → Elegiste: Ver perfil")
    elif opcion == "2":
        print("  → Elegiste: Ver notas")
    elif opcion == "3":
        print("  → Elegiste: Ver horario")
    elif opcion == "salir":
        print("  → Saliendo del sistema...")

print("Programa cerrado.")
````

### 🤔 ¿Cuándo usar `for` y cuándo usar `while`?

| | `for` | `while` |
|---|---|---|
| **Cuándo usarlo** | Cuando sabes cuántas veces repetir | Cuando repites hasta una condición |
| **Iterable necesario** | Sí | No |
| **Riesgo de infinito** | No | Sí, si no actualizas la condición |
| **Ejemplo típico** | Recorrer lista, tabla de multiplicar | Esperar input, juegos, simulaciones |

## 6. Palabras clave: `break`, `continue`, `else` 🎛️ <a id='6'></a>

### `break` — Salir del bucle inmediatamente

````python
# break: detiene el bucle al encontrar el primer número negativo
numeros = [4, 7, 2, -1, 9, 3, -5, 8]

print("Buscando el primer número negativo...")
for num in numeros:
    if num < 0:
        print(f"¡Encontrado! El primer negativo es: {num}")
        break
    print(f"  {num} es positivo, continúo...")
````

### `continue` — Saltar esta vuelta y seguir con la siguiente

````python
# continue: salta los números pares, solo imprime impares
print("Números impares del 1 al 10:")
for i in range(1, 11):
    if i % 2 == 0:   # si es par...
        continue     # ...salta esta vuelta
    print(i, end=" ")
````

### `else` — Se ejecuta si el bucle terminó sin `break`

````python
# else en un for: se ejecuta si no hubo break
def buscar_nombre(lista, objetivo):
    for nombre in lista:
        if nombre == objetivo:
            print(f"✅ '{objetivo}' encontrado en la lista.")
            break
    else:
        print(f"❌ '{objetivo}' NO está en la lista.")

lista_clase = ["Ana", "Luis", "Pedro", "Sofía", "María"]

buscar_nombre(lista_clase, "Pedro")   # existe
buscar_nombre(lista_clase, "Carlos")  # no existe
````

## 7. Bucles anidados 🪆 <a id='7'></a>

Un bucle **dentro** de otro bucle. Por cada vuelta del exterior, el interior completa todas sus vueltas.

````python
# Tabla de multiplicar completa (del 1 al 5)
print("📊 Tabla de multiplicar (1 al 5)\n")

# Encabezado
print("   ", end="")
for j in range(1, 6):
    print(f"  {j:2}", end="")
print()
print("   " + "----" * 5)

# Filas
for i in range(1, 6):
    print(f"{i:2} |", end="")
    for j in range(1, 6):
        print(f"  {i*j:2}", end="")
    print()  # nueva línea por fila
````

````python
# Patrón visual con bucles anidados
print("Triángulo de estrellas:")
for fila in range(1, 7):
    for columna in range(fila):
        print("★", end=" ")
    print()
````

````python
# Recorrer una lista de listas (matriz)
notas_clase = [
    ["Ana",   6.5, 7.0, 6.8],
    ["Luis",  5.5, 6.0, 5.8],
    ["Pedro", 7.0, 7.5, 6.9],
]

print("📋 Promedios por alumno:")
for fila in notas_clase:
    nombre = fila[0]
    notas  = fila[1:]
    promedio = sum(notas) / len(notas)
    print(f"  {nombre}: {promedio:.2f} → {'✅' if promedio >= 6.0 else '❌'}")
````

## 8. Casos reales y ejercicios finales 🚀 <a id='8'></a>

````python
# 🛒 CASO 1: Carrito de compras
carrito = [
    {"producto": "Laptop",   "precio": 500_000, "cantidad": 1},
    {"producto": "Mouse",    "precio": 15_000,  "cantidad": 2},
    {"producto": "Teclado",  "precio": 25_000,  "cantidad": 1},
    {"producto": "Monitor",  "precio": 180_000, "cantidad": 1},
]

print("🛒 Resumen del carrito:\n")
total = 0
for item in carrito:
    subtotal = item["precio"] * item["cantidad"]
    total += subtotal
    print(f"  {item['producto']:12} x{item['cantidad']}  ${subtotal:>10,.0f}")

print(f"  {'─'*35}")
print(f"  {'TOTAL':12}      ${total:>10,.0f}")
````

````python
# 🔐 CASO 2: Validador de contraseña (while)
def validar_password(password):
    """Valida que una contraseña cumpla requisitos mínimos"""
    errores = []
    
    if len(password) < 8:
        errores.append("❌ Debe tener al menos 8 caracteres")
    
    tiene_mayuscula = False
    tiene_numero = False
    
    for char in password:
        if char.isupper():
            tiene_mayuscula = True
        if char.isdigit():
            tiene_numero = True
    
    if not tiene_mayuscula:
        errores.append("❌ Debe tener al menos una letra mayúscula")
    if not tiene_numero:
        errores.append("❌ Debe tener al menos un número")
    
    return errores

# Probar con varias contraseñas
passwords = ["abc", "abcdefgh", "Abcdefgh", "Abcdef12"]

for pw in passwords:
    errores = validar_password(pw)
    if errores:
        print(f"\nPassword '{pw}' — inválida:")
        for error in errores:
            print(f"  {error}")
    else:
        print(f"\nPassword '{pw}' — ✅ Válida!")
````

````python
# 📊 CASO 3: Estadísticas de temperatura
temperaturas = [18.5, 21.0, 19.5, 23.0, 17.5, 22.5, 20.0, 24.5, 18.0, 21.5]

# Calcular estadísticas con bucle
suma = 0
minima = temperaturas[0]
maxima = temperaturas[0]
dias_calurosos = 0  # más de 22°C

for temp in temperaturas:
    suma += temp
    if temp < minima:
        minima = temp
    if temp > maxima:
        maxima = temp
    if temp > 22:
        dias_calurosos += 1

promedio = suma / len(temperaturas)

print("🌡️  Estadísticas de temperatura:")
print(f"   Mínima:          {minima}°C")
print(f"   Máxima:          {maxima}°C")
print(f"   Promedio:        {promedio:.1f}°C")
print(f"   Días calurosos:  {dias_calurosos} de {len(temperaturas)}")
````

## 📚 Resumen General

| Concepto | Sintaxis | Cuándo usarlo |
|----------|----------|---------------|
| `for` básico | `for x in lista:` | Recorrer una secuencia |
| `for` con range | `for i in range(n):` | Repetir n veces |
| `enumerate` | `for i, v in enumerate(lista):` | Necesitas índice y valor |
| `zip` | `for a, b in zip(l1, l2):` | Recorrer dos listas juntas |
| `while` | `while condicion:` | Repetir hasta condición |
| `break` | dentro del bucle | Salir del bucle antes de terminar |
| `continue` | dentro del bucle | Saltar la vuelta actual |
| `else` | después del bucle | Ejecutar si no hubo `break` |
| Anidados | `for` dentro de `for` | Matrices, tablas, patrones |
